```python
import requests
from bs4 import BeautifulSoup as bs
import lxml
```

```python
res1 = requests.get('https://www.w3schools.com/css/default.asp')
res2 = requests.get('https://www.youtube.com/watch?v=GI1XuNB2wgE')
```

```python
res1.text
```

```python
soup = bs(res1.text,'lxml')
soup
```

```python
soup.select('title')
```

```python
list_of_div = soup.select('div')
len(list_of_div)
```
